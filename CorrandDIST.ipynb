```python
df = pd.read_csv("C:\\CAPSJUPYTER\\Crop_recommendation.csv")
```

```python
# Correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Features", fontsize=14)
plt.show()
```

```python
# Distribution plots
df.hist(bins=20, figsize=(12,8), color='orange', edgecolor='black')
plt.suptitle("Distribution of Dataset Features", fontsize=16)
plt.tight_layout()
plt.show()
```
